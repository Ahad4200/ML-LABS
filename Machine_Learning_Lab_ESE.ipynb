## Compute and visualize silhouette score

### Subtask:
Compute the silhouette score and visualize using a silhouette plot.


**Reasoning**:
Import the necessary metrics functions for calculating silhouette scores and samples.



```python
from sklearn.metrics import silhouette_score, silhouette_samples
```

**Reasoning**:
Compute and print the silhouette score for the clustering result on the noisy dataset.



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score, silhouette_samples
import matplotlib.cm as cm # Import the cm module

unique_noisy_labels = np.unique(noisy_labels)
# Exclude the noise label (-1) for silhouette score calculation
cluster_labels = unique_noisy_labels[unique_noisy_labels != -1]

if len(cluster_labels) < 2:
    print("Silhouette score cannot be computed: Less than 2 clusters found (excluding noise).")
else:
    # Compute the average silhouette score
    silhouette_avg = silhouette_score(X_noisy, noisy_labels)
    print(f"Average Silhouette Score: {silhouette_avg}")

    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(X_noisy, noisy_labels)

    # Generate the silhouette plot
    fig, ax = plt.subplots(1, 1, figsize=(8, 6))

    y_lower = 10
    for i in cluster_labels:
        # Aggregate the silhouette scores for samples belonging to cluster i, and sort them
        ith_cluster_silhouette_values = sample_silhouette_values[noisy_labels == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / len(cluster_labels)) # Use cm.nipy_spectral
        ax.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax.set_xlabel("Silhouette coefficient values")
    ax.set_ylabel("Cluster label")
    ax.set_title("Silhouette Plot for DBSCAN Clustering")

    # The vertical line for average silhouette score of all the values
    ax.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax.set_yticks([])  # Clear the y-axis labels / ticks
    ax.set_xticks(np.arange(-1, 1.1, 0.1))

    plt.show()
```

```python
from sklearn.datasets import make_blobs
import time

def stream_data(n_samples=1000, centers=3, cluster_std=0.5, random_state=42, delay=0.1):
    """Simulates a continuous stream of 2D data points."""
    X, _ = make_blobs(n_samples=n_samples, centers=centers, cluster_std=cluster_std, random_state=random_state)
    for point in X:
        yield point
        time.sleep(delay) # Simulate real-time delay

# Example usage (will be used later in the real-time processing)
data_stream = stream_data()
for i, point in enumerate(data_stream):
  print(f"Received point {i}: {point}")
  if i > 5: # Stop after a few points for demonstration
    break
```

### Euclidean Distance Calculation
Implement a function to calculate the Euclidean distance between two points.

**Reasoning**:
Implement a function to calculate the Euclidean distance using numpy's `linalg.norm`.

```python
import numpy as np

def euclidean_distance(point1, point2):
    """Calculates the Euclidean distance between two points."""
    return np.linalg.norm(point1 - point2)

# Example usage:
point_a = np.array([1, 2])
point_b = np.array([4, 6])
distance = euclidean_distance(point_a, point_b)
print(f"Euclidean distance between {point_a} and {point_b}: {distance}")
```

### Data Point Assignment
Assign incoming data points to the nearest cluster centroid.

**Reasoning**:
Implement a function to assign a data point to the nearest cluster centroid by calculating the distance to each centroid and selecting the closest one.

```python
def assign_to_nearest_centroid(point, centroids):
    """Assigns a data point to the nearest cluster centroid."""
    distances = [euclidean_distance(point, centroid) for centroid in centroids]
    nearest_centroid_index = np.argmin(distances)
    return nearest_centroid_index

# Example usage (requires centroids to be defined):
centroids = [np.array([0, 0]), np.array([5, 5])]
new_point = np.array([1, 1])
assigned_cluster = assign_to_nearest_centroid(new_point, centroids)
print(f"Point {new_point} assigned to cluster {assigned_cluster}")
```

### Online K-Means Clustering
Update centroids dynamically as new points arrive.

**Reasoning**:
Implement a class for Online K-Means clustering that initializes centroids and updates them incrementally with each new data point.

```python
class OnlineKMeans:
    def __init__(self, n_clusters, random_state=42):
        self.n_clusters = n_clusters
        self.centroids = None
        self.cluster_counts = None
        self.random_state = random_state
        np.random.seed(random_state)

    def initialize_centroids(self, data_point):
        """Initializes centroids with the first data points."""
        if self.centroids is None:
            self.centroids = np.array([data_point])
            self.cluster_counts = np.array([1])
        elif len(self.centroids) < self.n_clusters:
             # Check if the point is already a centroid to avoid duplicates
            if not any(np.array_equal(data_point, centroid) for centroid in self.centroids):
                self.centroids = np.vstack([self.centroids, data_point])
                self.cluster_counts = np.append(self.cluster_counts, 1)


    def update_centroid(self, point, assigned_cluster_index):
        """Updates the centroid of the assigned cluster incrementally."""
        # Increase the count of points in the assigned cluster
        self.cluster_counts[assigned_cluster_index] += 1
        n = self.cluster_counts[assigned_cluster_index]
        # Update the centroid using the formula: new_centroid = old_centroid + (new_point - old_centroid) / n
        self.centroids[assigned_cluster_index] += (point - self.centroids[assigned_cluster_index]) / n

    def process_point(self, point):
        """Processes a new data point: initializes centroids or assigns and updates."""
        if self.centroids is None or len(self.centroids) < self.n_clusters:
            self.initialize_centroids(point)
            assigned_cluster_index = None # No assignment until all centroids are initialized
        else:
            assigned_cluster_index = assign_to_nearest_centroid(point, self.centroids)
            self.update_centroid(point, assigned_cluster_index)
        return assigned_cluster_index

# Example usage (will be used in the real-time processing loop):
online_kmeans = OnlineKMeans(n_clusters=3)
# Simulate processing a few points
for i in range(10):
  point = np.random.rand(2) * 10 # Generate a random 2D point
  assigned_cluster = online_kmeans.process_point(point)
  if online_kmeans.centroids is not None:
    print(f"Processed point {point}, assigned to cluster {assigned_cluster}, Centroids: {online_kmeans.centroids}")
```

### Outlier Detection
Identify outliers using a distance threshold.

**Reasoning**:
Implement a function within the `OnlineKMeans` class to identify outliers based on their distance to the nearest centroid and a predefined threshold.

```python
class OnlineKMeans:
    def __init__(self, n_clusters, outlier_threshold=1.5, random_state=42):
        self.n_clusters = n_clusters
        self.centroids = None
        self.cluster_counts = None
        self.outlier_threshold = outlier_threshold # Threshold for outlier detection
        self.random_state = random_state
        np.random.seed(random_state)

    def initialize_centroids(self, data_point):
        """Initializes centroids with the first data points."""
        if self.centroids is None:
            self.centroids = np.array([data_point])
            self.cluster_counts = np.array([1])
        elif len(self.centroids) < self.n_clusters:
             # Check if the point is already a centroid to avoid duplicates
            if not any(np.array_equal(data_point, centroid) for centroid in self.centroids):
                self.centroids = np.vstack([self.centroids, data_point])
                self.cluster_counts = np.append(self.cluster_counts, 1)


    def update_centroid(self, point, assigned_cluster_index):
        """Updates the centroid of the assigned cluster incrementally."""
        # Increase the count of points in the assigned cluster
        self.cluster_counts[assigned_cluster_index] += 1
        n = self.cluster_counts[assigned_cluster_index]
        # Update the centroid using the formula: new_centroid = old_centroid + (new_point - old_centroid) / n
        self.centroids[assigned_cluster_index] += (point - self.centroids[assigned_cluster_index]) / n

    def is_outlier(self, point):
        """Checks if a data point is an outlier based on distance to nearest centroid."""
        if self.centroids is None or len(self.centroids) < self.n_clusters:
            return False # Cannot determine outliers until all centroids are initialized

        # Find the distance to the nearest centroid
        distances = [euclidean_distance(point, centroid) for centroid in self.centroids]
        min_distance = np.min(distances)

        # Check if the minimum distance exceeds the outlier threshold
        return min_distance > self.outlier_threshold


    def process_point(self, point):
        """Processes a new data point: initializes centroids, checks for outliers, or assigns and updates."""
        if self.centroids is None or len(self.centroids) < self.n_clusters:
            self.initialize_centroids(point)
            assigned_cluster_index = None # No assignment until all centroids are initialized
            is_outlier_flag = False # Cannot be an outlier if centroids are not initialized
        else:
            is_outlier_flag = self.is_outlier(point)
            if is_outlier_flag:
                assigned_cluster_index = -1 # Assign -1 label for outliers
            else:
                assigned_cluster_index = assign_to_nearest_centroid(point, self.centroids)
                self.update_centroid(point, assigned_cluster_index)

        return assigned_cluster_index, is_outlier_flag

# Example usage (will be used in the real-time processing loop):
online_kmeans = OnlineKMeans(n_clusters=3, outlier_threshold=2.0) # Example threshold
# Simulate processing a few points
for i in range(10):
  point = np.random.rand(2) * 10 # Generate a random 2D point
  assigned_cluster, is_outlier = online_kmeans.process_point(point)
  if online_kmeans.centroids is not None:
    print(f"Processed point {point}, assigned to cluster {assigned_cluster}, is outlier: {is_outlier}, Centroids: {online_kmeans.centroids}")
```

### Micro-Cluster Summarization
Maintain and update cluster statistics like count, mean, and variance.

**Reasoning**:
Enhance the `OnlineKMeans` class to maintain and update the sum and sum of squares for each cluster, which can be used to calculate mean and variance.

```python
class OnlineKMeans:
    def __init__(self, n_clusters, outlier_threshold=1.5, random_state=42):
        self.n_clusters = n_clusters
        self.centroids = None
        self.cluster_counts = None
        self.cluster_sums = None # To store the sum of points in each cluster
        self.cluster_sum_sqs = None # To store the sum of squares of points in each cluster
        self.outlier_threshold = outlier_threshold
        self.random_state = random_state
        np.random.seed(random_state)

    def initialize_centroids(self, data_point):
        """Initializes centroids with the first data points."""
        if self.centroids is None:
            self.centroids = np.array([data_point])
            self.cluster_counts = np.array([1])
            self.cluster_sums = np.array([data_point.copy()]) # Initialize sum with the first point
            self.cluster_sum_sqs = np.array([(data_point**2).copy()]) # Initialize sum of squares
        elif len(self.centroids) < self.n_clusters:
             # Check if the point is already a centroid to avoid duplicates
            if not any(np.array_equal(data_point, centroid) for centroid in self.centroids):
                self.centroids = np.vstack([self.centroids, data_point])
                self.cluster_counts = np.append(self.cluster_counts, 1)
                self.cluster_sums = np.vstack([self.cluster_sums, data_point.copy()]) # Add sum for the new cluster
                self.cluster_sum_sqs = np.vstack([self.cluster_sum_sqs, (data_point**2).copy()]) # Add sum of squares for the new cluster


    def update_centroid(self, point, assigned_cluster_index):
        """Updates the centroid and statistics of the assigned cluster incrementally."""
        # Update sum and sum of squares
        self.cluster_sums[assigned_cluster_index] += point
        self.cluster_sum_sqs[assigned_cluster_index] += point**2

        # Increase the count of points in the assigned cluster
        self.cluster_counts[assigned_cluster_index] += 1
        n = self.cluster_counts[assigned_cluster_index]

        # Update the centroid using the formula: new_centroid = old_centroid + (new_point - old_centroid) / n
        self.centroids[assigned_cluster_index] = self.cluster_sums[assigned_cluster_index] / n


    def is_outlier(self, point):
        """Checks if a data point is an outlier based on distance to nearest centroid."""
        if self.centroids is None or len(self.centroids) < self.n_clusters:
            return False

        distances = [euclidean_distance(point, centroid) for centroid in self.centroids]
        min_distance = np.min(distances)

        return min_distance > self.outlier_threshold

    def get_cluster_stats(self):
        """Returns the mean and variance for each cluster."""
        stats = {}
        if self.centroids is not None and len(self.centroids) == self.n_clusters:
            for i in range(self.n_clusters):
                n = self.cluster_counts[i]
                mean = self.cluster_sums[i] / n
                # Variance = (Sum of squares / n) - (Mean)^2
                variance = (self.cluster_sum_sqs[i] / n) - (mean**2)
                stats[f'cluster_{i}'] = {'count': n, 'mean': mean, 'variance': variance}
        return stats


    def process_point(self, point):
        """Processes a new data point: initializes centroids, checks for outliers, or assigns and updates."""
        if self.centroids is None or len(self.centroids) < self.n_clusters:
            self.initialize_centroids(point)
            assigned_cluster_index = None
            is_outlier_flag = False
        else:
            is_outlier_flag = self.is_outlier(point)
            if is_outlier_flag:
                assigned_cluster_index = -1 # Assign -1 label for outliers
            else:
                assigned_cluster_index = assign_to_nearest_centroid(point, self.centroids)
                self.update_centroid(point, assigned_cluster_index)

        return assigned_cluster_index, is_outlier_flag

# Example usage (will be used in the real-time processing loop):
online_kmeans = OnlineKMeans(n_clusters=3, outlier_threshold=2.0)
# Simulate processing a few points
for i in range(20): # Process more points to get meaningful stats
  point = np.random.rand(2) * 10
  assigned_cluster, is_outlier = online_kmeans.process_point(point)
  if online_kmeans.centroids is not None and len(online_kmeans.centroids) == online_kmeans.n_clusters:
    print(f"Processed point {point}, assigned to cluster {assigned_cluster}, is outlier: {is_outlier}")

if online_kmeans.centroids is not None and len(online_kmeans.centroids) == online_kmeans.n_clusters:
  cluster_stats = online_kmeans.get_cluster_stats()
  print("\nCluster Statistics:")
  for cluster, stats in cluster_stats.items():
    print(f"{cluster}: Count = {stats['count']}, Mean = {stats['mean']}, Variance = {stats['variance']}")
```

### Concept Drift Detection
Monitor cluster sizes and detect concept drift based on significant changes.

**Reasoning**:
Add a method to the `OnlineKMeans` class to detect concept drift by monitoring changes in cluster counts relative to the total number of processed points.

```python
class OnlineKMeans:
    def __init__(self, n_clusters, outlier_threshold=1.5, concept_drift_threshold=0.1, random_state=42):
        self.n_clusters = n_clusters
        self.centroids = None
        self.cluster_counts = None
        self.cluster_sums = None
        self.cluster_sum_sqs = None
        self.outlier_threshold = outlier_threshold
        self.concept_drift_threshold = concept_drift_threshold # Threshold for concept drift detection
        self.total_points_processed = 0
        self.random_state = random_state
        np.random.seed(random_state)

    def initialize_centroids(self, data_point):
        """Initializes centroids with the first data points."""
        if self.centroids is None:
            self.centroids = np.array([data_point])
            self.cluster_counts = np.array([1])
            self.cluster_sums = np.array([data_point.copy()])
            self.cluster_sum_sqs = np.array([(data_point**2).copy()])
        elif len(self.centroids) < self.n_clusters:
             # Check if the point is already a centroid to avoid duplicates
            if not any(np.array_equal(data_point, centroid) for centroid in self.centroids):
                self.centroids = np.vstack([self.centroids, data_point])
                self.cluster_counts = np.append(self.cluster_counts, 1)
                self.cluster_sums = np.vstack([self.cluster_sums, data_point.copy()])
                self.cluster_sum_sqs = np.vstack([self.cluster_sum_sqs, (data_point**2).copy()])


    def update_centroid(self, point, assigned_cluster_index):
        """Updates the centroid and statistics of the assigned cluster incrementally."""
        self.cluster_sums[assigned_cluster_index] += point
        self.cluster_sum_sqs[assigned_cluster_index] += point**2
        self.cluster_counts[assigned_cluster_index] += 1
        n = self.cluster_counts[assigned_cluster_index]
        self.centroids[assigned_cluster_index] = self.cluster_sums[assigned_cluster_index] / n

    def is_outlier(self, point):
        """Checks if a data point is an outlier based on distance to nearest centroid."""
        if self.centroids is None or len(self.centroids) < self.n_clusters:
            return False

        distances = [euclidean_distance(point, centroid) for centroid in self.centroids]
        min_distance = np.min(distances)

        return min_distance > self.outlier_threshold

    def detect_concept_drift(self):
        """Detects concept drift based on significant changes in cluster sizes."""
        if self.centroids is None or len(self.centroids) < self.n_clusters or self.total_points_processed < 100: # Add a minimum points check
            return False

        # Calculate the proportion of points in each cluster
        cluster_proportions = self.cluster_counts / self.total_points_processed


        average_cluster_count = self.total_points_processed / self.n_clusters
        for count in self.cluster_counts:
            if count < average_cluster_count * (1 - self.concept_drift_threshold):
                 print(f"Potential concept drift detected: Cluster count {count} is significantly low compared to average {average_cluster_count:.2f}")
                 return True
        return False


    def get_cluster_stats(self):
        """Returns the mean and variance for each cluster."""
        stats = {}
        if self.centroids is not None and len(self.centroids) == self.n_clusters:
            for i in range(self.n_clusters):
                n = self.cluster_counts[i]
                mean = self.cluster_sums[i] / n
                variance = (self.cluster_sum_sqs[i] / n) - (mean**2)
                stats[f'cluster_{i}'] = {'count': n, 'mean': mean, 'variance': variance}
        return stats


    def process_point(self, point):
        """Processes a new data point: initializes centroids, checks for outliers, or assigns and updates."""
        self.total_points_processed += 1
        if self.centroids is None or len(self.centroids) < self.n_clusters:
            self.initialize_centroids(point)
            assigned_cluster_index = None
            is_outlier_flag = False
        else:
            is_outlier_flag = self.is_outlier(point)
            if is_outlier_flag:
                assigned_cluster_index = -1 # Assign -1 label for outliers
            else:
                assigned_cluster_index = assign_to_nearest_centroid(point, self.centroids)
                self.update_centroid(point, assigned_cluster_index)

        if self.total_points_processed > self.n_clusters and self.total_points_processed % 100 == 0:
             self.detect_concept_drift()


        return assigned_cluster_index, is_outlier_flag

# Example usage (will be used in the real-time processing loop):
online_kmeans = OnlineKMeans(n_clusters=3, outlier_threshold=2.0, concept_drift_threshold=0.2) # Example thresholds
# Simulate processing a few points
for i in range(300): # Process more points to potentially see concept drift
     point = np.random.rand(2) * 10
     # Simulate a concept drift after 150 points by changing the data distribution
     if i > 150:
          point = np.random.rand(2) * 5 + 15 # New cluster appearing

     assigned_cluster, is_outlier = online_kmeans.process_point(point)
     if online_kmeans.centroids is not None and len(online_kmeans.centroids) == online_kmeans.n_clusters:
       print(f"Processed point {point}, assigned to cluster {assigned_cluster}, is outlier: {is_outlier}")

if online_kmeans.centroids is not None and len(online_kmeans.centroids) == online_kmeans.n_clusters:
  cluster_stats = online_kmeans.get_cluster_stats()
  print("\nCluster Statistics:")
  for cluster, stats in cluster_stats.items():
    print(f"{cluster}: Count = {stats['count']}, Mean = {stats['mean']}, Variance = {stats['variance']}")
```
